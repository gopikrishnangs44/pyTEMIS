# USAGE

Monthly HCHO and NO2 data is available through https://www.temis.nl/index.php


Available options are

1.  pyTEMIS.TEMIS_dat_GOME2A_HCHO    ##GOME2A HCHO 
2.  pyTEMIS.TEMIS_dat_GOME2B_HCHO    ##GOME2B HCHO
3.  pyTEMIS.TEMIS_dat_GOME_HCHO      ##GOME HCHO
4.  pyTEMIS.TEMIS_dat_OMI_HCHO       ##OMI HCHO
5.  pyTEMIS.TEMIS_dat_SCIA_HCHO      ##SCIAMACHY HCHO
6.  pyTEMIS.TEMIS_GRD_GOME2A_NO2     ##GOME2A NO2
7.  pyTEMIS.TEMIS_GRD_GOME2B_NO2     ##GOME2B NO2
8.  pyTEMIS.TEMIS_GRD_OMI_NO2        ##GOME NO2
9.  pyTEMIS.TEMIS_GRD_SCIA_NO2       ##SCIAMACHY NO2
10. pyTEMIS.TEMIS_TOMS_TROPOMI_NO2   ##TROPOMI NO2


TEMIS_DAT indicates that the data should be downloaded in .dat format

TEMIS_GRD indicates that the data should be downloaded in .grd format

TEMIS_TOMS indicates that the data should be downloaded in .asc format


*******************************************************************************
Units are not converted in this package. It will be mentioned in the raw data

for eg:

; GOME2 CH2O H2CO_VCD_w320 Monthly Averaged (x 1e15 [molec/cm2]) 

; FRESCOv6 cloud fractions <= 40%  

; SZA <= 60  

; 01/2013 

; Data gridded on 0.25 x 0.25 (1440 columns-longitudes, 720 lines-latitudes) 

; Longitudes from -179.88 W to 179.88 E.

; Latitudes from -89.88 S to 89.88 N 

*******************************************************************************

```python
pyTEMIS.TEMIS_dat_GOME2A_HCHO('GOME2CH2O_Grid_720x1440_200701.dat').plot(cmap='jet', vmin=0, vmax=30)
```
